```python
#Qlearning
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from matplotlib.patches import Patch
import glob

# Load all policy files
file_paths = [
"policies\\hard\\qtable.csv",
"policies\\hard\\dqn.csv",
"policies\\hard\\optimal.csv"
]
policies = [pd.read_csv(f, index_col=0) for f in file_paths]
print(file_paths)

titles = ['Q-Learning', 'DQN', 'Optimal']  # Modify these titles

# Set up plot parameters
plt.rcParams.update({
    'font.size': 10,
    'axes.titlesize': 12,
    'axes.labelsize': 10,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'font.family': 'serif',
    'figure.dpi': 300
})

# Create figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 4.5), sharey=True, 
                        constrained_layout=True)
fig.suptitle("Hard Hands", 
            x=0.525, y=1.06, fontsize=14, fontweight='bold')

cmap = mcolors.ListedColormap(['#d62728', '#2ca02c'])  # Consistent color mapping

for ax, policy, title in zip(axes, policies, titles):
    annot_df = policy.replace({0.0: 'S', 1.0: 'H'})
    
    sns.heatmap(
        policy,
        cmap=cmap,
        annot=annot_df,
        fmt='',
        cbar=False,
        linewidths=0.3,
        linecolor='white',
        annot_kws={'fontsize': 8, 'color': 'white', 'weight': 'bold'},
        ax=ax
    )
    
    # Axis formatting
    ax.set_xlabel("Dealer Upcard", labelpad=8)
    ax.set_ylabel("Player Total", labelpad=8) if ax == axes[0] else None
    ax.set_title(title, pad=10, fontsize=11)
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')

# Create unified legend below all plots
legend_elements = [
    Patch(facecolor='#2ca02c', edgecolor='white', label='Stand (S)'),
    Patch(facecolor='#d62728', edgecolor='white', label='Hit (H)')
]

fig.legend(
    handles=legend_elements,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=2,
    frameon=False,
    fontsize=10
)


# Adjust layout and save
plt.savefig('combined_strategies.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
